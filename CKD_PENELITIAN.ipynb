```python
import pandas as pd

url = "https://raw.githubusercontent.com/Putrimaesarah/CKD-DATASET/main/Chronic_Kidney_Dsease_data.csv"
df = pd.read_csv(url)

# Cek data awal
print(df.shape)
df.head()
```

```python


# Menampilkan ukuran data
print("Jumlah baris dan kolom:", df.shape)

```

```python
# Mengecek jumlah missing value di setiap kolom
print("\nJumlah missing value per kolom:")
print(df.isnull().sum())
```

```python
# Opsional: Menampilkan kolom yang memiliki missing value saja
print("\nKolom yang memiliki missing value:")
missing_cols = df.columns[df.isnull().any()]
print(missing_cols)
```

```python
# Hitung jumlah pasien berdasarkan Diagnosis
jumlah_diagnosis = df['Diagnosis'].value_counts()

print("Jumlah pasien berdasarkan Diagnosis:")
print(jumlah_diagnosis)
```

```python
# Melihat tipe data tiap kolom
df.dtypes
```

```python
# Menampilkan tipe data dan jumlah unique value per kolom (jika ingin analisis lebih lanjut)
data_types = pd.DataFrame({
    'Tipe Data': df.dtypes,
    'Jumlah Nilai Unik': df.nunique()
})
data_types
```

```python
# Pisahkan kolom numerik dan kategorikal
kolom_numerik = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
kolom_kategorikal = df.select_dtypes(include=['object']).columns.tolist()

print("Jumlah Kolom Numerik:", len(kolom_numerik))
print("Kolom Numerik:")
print(kolom_numerik)

print("\nJumlah Kolom Kategorikal:", len(kolom_kategorikal))
print("Kolom Kategorikal:")
print(kolom_kategorikal)
```

```python
#HAPUS KOLOM
df.drop(columns=['PatientID', 'DoctorInCharge'], inplace=True)
```

```python
# Opsional: konversi ke kategori
categorical_cols = ['Gender', 'Ethnicity', 'SocioeconomicStatus', 'EducationLevel',
                    'Smoking', 'FamilyHistoryKidneyDisease', 'FamilyHistoryHypertension',
                    'FamilyHistoryDiabetes', 'PreviousAcuteKidneyInjury',
                    'UrinaryTractInfections', 'ACEInhibitors', 'Diuretics', 'Statins',
                    'AntidiabeticMedications', 'Edema', 'HeavyMetalsExposure',
                    'OccupationalExposureChemicals', 'WaterQuality']

for col in categorical_cols:
    df[col] = df[col].astype('category')
```

```python
# Menampilkan tipe data dan jumlah unique value per kolom (jika ingin analisis lebih lanjut)
data_types = pd.DataFrame({
    'Tipe Data': df.dtypes,
    'Jumlah Nilai Unik': df.nunique()
})
data_types
```
